```python
cols = [ 'MonthlyMaximumTemperature', 'MonthlyMinimumTemperature', 'MonthlyMeanTemperature','MonthlyTotalLiquidPrecipitation', 'MonthlyGreatestPrecip', 'MonthlyGreatestSnowfall']
plt.figure(figsize=(12, 8))
for i, column in enumerate(cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(monthly[column], kde=True)
    plt.title(column)
    plt.xlabel(column)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x="DATE", y="MonthlyMeanTemperature", data=monthly, color = "BROWN")
plt.title("Monthly Mean Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature in Celsius")
plt.xticks(rotation=45)
plt.show()
```

```python
attr = [
 'AWND',
 'CDSD',
 'CLDD',
 'DSNW',
 'HDSD',
 'HTDD',
 'MonthlyDaysWithGT001Precip',
 'MonthlyDaysWithGT010Precip',
 'MonthlyDaysWithGT32Temp',
 'MonthlyDaysWithGT90Temp',
 'MonthlyDaysWithLT0Temp',
 'MonthlyDaysWithLT32Temp',
 'MonthlyDepartureFromNormalAverageTemperature',
 'MonthlyDepartureFromNormalCoolingDegreeDays',
 'MonthlyDepartureFromNormalHeatingDegreeDays',
 'MonthlyDepartureFromNormalMaximumTemperature',
 'MonthlyDepartureFromNormalMinimumTemperature',
 'MonthlyDepartureFromNormalPrecipitation',
 'MonthlyGreatestPrecip',
 'MonthlyMaximumTemperature',
 'MonthlyMeanTemperature',
 'MonthlyMinimumTemperature',
 'MonthlySeaLevelPressure',
 'MonthlyTotalLiquidPrecipitation',
 'NormalsHeatingDegreeDay']
monthly[attr] = monthly[attr].apply(pd.to_numeric, errors = 'coerce')
corr_mat = monthly[attr].corr()
plt.figure(figsize=(16,12))
sns.heatmap(corr_mat, annot = True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap")
plt.show()
```

```python
colour = ['skyblue', 'green', 'pink', 'purple', 'orange','coral']
plot_cols = ['MonthlyMeanTemperature',
 'MonthlyMinimumTemperature','MonthlyMaximumTemperature','MonthlyGreatestPrecip','MonthlySeaLevelPressure', 'MonthlyGreatestSnowDepth']
for i,(column,color) in enumerate(zip(plot_cols, colour), start=1):
    plt.figure(figsize=(16,12))
    plt.subplot(3,2,i)
    plt.hist(monthly[column], color = color, bins=20)
    plt.title(column)
    plt.xlabel(column.split('Monthly',1)[-1].split('Value',1)[-1])
    plt.ylabel('Frequency')
    plt.tight_layout
plt.show()
```

```python
pairplot_columns = monthly[['AWND', 'CDSD', 'CLDD', 'DSNW', 'HDSD', 'HTDD']]
sns.pairplot(pairplot_columns)
plt.suptitle('Pair Plot for Numerical Variables', y=1.00)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = ['MonthlyMinimumTemperature','MonthlyMaximumTemperature','MonthlyMeanTemperature','MonthlySeaLevelPressure','MonthlyGreatestPrecip']
for i in decompose:
    print(i)
    res = seasonal_decompose(monthly[i], model='additive', period=12)
    plt.figure(figsize = (15,10))
    plt.subplot(4,1,1)
    plt.plot(res.trend)
    plt.title('Trend Component')
    plt.subplot(4,1,2)
    plt.plot(res.seasonal)
    plt.title('Seasonal Component')
    plt.subplot(4,1,3)
    plt.plot(res.resid)
    plt.title('Residual Component')
    plt.subplot(4,1,4)
    plt.plot(res.observed)
    plt.title('Observed Data')
    plt.tight_layout()
    plt.show()
```

```python
# Convert 'DATE' column to datetime format
monthly['DATE'] = pd.to_datetime(monthly['DATE'])
monthly.set_index('DATE', inplace=True)
cols =  ['MonthlyMinimumTemperature','MonthlyMaximumTemperature','MonthlyMeanTemperature','MonthlySeaLevelPressure','MonthlyGreatestPrecip']
for i in cols:
    plt.figure(figsize=(15, 6))
    plt.plot(monthly[i])
    plt.title('Time Series Analysis - '+i)
    plt.xlabel('Date')
    plt.ylabel('Climate Feature')
    plt.show()
```

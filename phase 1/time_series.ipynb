```python
plt.figure(figsize=(16,8))
plt.plot(data['Date'], data['AMZN'], color='blue')
plt.title('AMZN - stock predictions')
plt.xlabel('Dates')
plt.ylabel('Stock Value')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.plot(data['Date'], data['DPZ'], color='blue')
plt.title('DPZ - stock predictions')
plt.xlabel('Dates')
plt.ylabel('Stock Value')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.plot(data['Date'], data['NFLX'], color='blue')
plt.title('NFLX - stock predictions')
plt.xlabel('Dates')
plt.ylabel('Stock Value')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.plot(data['Date'], data['BTC'], color='blue')
plt.title('BTC - price predictions')
plt.xlabel('Dates')
plt.ylabel('Price')
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```

```python
year = pd.to_datetime(data['Date']).dt.year
stocks = data[['AMZN','DPZ','BTC','NFLX']].groupby(year, axis = 0).sum()
stocks
```

```python
data_year = data
data_year['Date'] = pd.to_datetime(data_year['Date']).dt.year
data_year
```

```python
plt.figure(figsize=(16,8))
stocks.plot()
plt.show()
```

```python
stock_avg = data[['AMZN','DPZ','BTC','NFLX']].groupby(year, axis = 0).mean()
stock_avg
```

```python
plt.figure(figsize = (16,8))
stock_avg.plot.bar()
plt.show()
```

```python
ave_x = data['Date'].mean()
ave_y = data['AMZN'].mean()
max_y = data['AMZN'].max()
min_y = data['AMZN'].min()

fig, ax = plt.subplots(figsize=(10,6), facecolor='#4bd659')
ax.plot(data['Date'], data['AMZN'], color='#283954')
ax.set_title("AMZN stock trade volume", fontsize=20)
ax.set_xlabel("Date", fontsize=15)
ax.set_ylabel("Daily trade volume", fontsize=15)
ax.axhline(y=data['AMZN'].max(), linestyle='--', lw=2.2, color='green')
ax.axhline(y=data['AMZN'].min(), linestyle='--',lw=2.2, color='red')
ax.axhline(y=data['AMZN'].mean(), linestyle='--',lw=2.8, color='yellow')
ax.axvline(x=data[data['AMZN'] == max_y]['Date'].values, ls='--', lw='2.2', color='#0aebff')
ax.annotate(f"Average AMZN Stock Price {round(data_year['AMZN'].mean(),2)}", 
                xy=(ave_x,ave_y),
                bbox=dict(boxstyle="round",facecolor='#e8e0ba', edgecolor='#d0d5db')
                 )

ax.annotate(f"Maximum AMZN Stock Price {data_year['AMZN'].max()}", 
                xy=(ave_x,max_y),
                bbox=dict(boxstyle="round",facecolor='#e8e0ba', edgecolor='#d0d5db')
                 )

ax.annotate(f"Minimum AMZN Stock Price {data_year['AMZN'].min()}", 
                xy=(ave_x,min_y),
                bbox=dict(boxstyle="round",facecolor='#e8e0ba', edgecolor='#d0d5db')
                 )
plt.show()
```
